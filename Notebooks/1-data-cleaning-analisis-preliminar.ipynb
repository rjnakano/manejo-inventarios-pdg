## Manejo de registros con nulos
* Eliminación de la columna IPG por su gran cantidad de valores vacios
* Crearemos dos data frames:
    * df_na_free: eliminará los registros nulos. 
    * df_imp: Reemplazará los NA con diferentes técnicas.
* Imputar valores aleatorios en las columnas HTS, TYPE, COLUMN, AISLE

```python
df.drop(labels=['IPG','SKU','LOC'],axis=1,inplace=True)
df.columns
```

```python
# Borramos cualquier fila con todos los valores nulos
df.dropna(axis=0,how="all", inplace=True)

# Creamos df_na_free eliminando los registros con valores NA
df_na_free = df.dropna(axis=0,how="any", inplace=False)
```

```python
df_imp = df

# Reemplazamos los valores NA del pasillo por el número 0
df_imp.AISLE.fillna(value=0,inplace=True)
df_imp.AISLE.describe()

# Reemplazamos los valores NA del pasillo por el texto N/A
df_imp.COLUMN.fillna(value='N/A', inplace=True)
df_imp.COLUMN.unique()

df_imp.TYPE.fillna(value='N/A', inplace=True)
df_imp.TYPE.unique()

df_imp.HTS.fillna(value='NO HTS', inplace=True)
df_imp.HTS.unique()
```

## Variables Dummy
Se crean dos dataframes nuevos con variables dummies, reemplazando las variables categóricas
* df_na_free_dummy: Basado en el dataframe que eliminó los registros vacíos
* df_imp_dummy: Basado en el dataframe que imputó los valores vacíos

```python
def get_dummied_df(df):
    df_dummies = df
    col_obj = list()
    for col in df.columns.values:
        if df[str(col)].dtype == 'object':
            col_obj.append(col)
            dummy = pd.get_dummies(df[col],prefix=col)
            df_dummies = pd.merge(df_dummies, dummy, left_index= True, right_index= True)

    df_dummies.drop(labels=col_obj, axis=1, inplace = True)
    return df_dummies
```

```python
df_na_free_dummy = get_dummied_df ( df_na_free )
df_imp_dummy = get_dummied_df ( df_imp )
```
